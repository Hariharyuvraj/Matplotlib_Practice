```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Pie Chart


```python
department=['Hr','production','maintenance','logistic','purchase']
budget=[1,10,5,8,4]
```

```python
plt.pie(budget,labels=department,autopct="%0.1f%%")
plt.show()
```

## Scatter Plot

```python
x=[10,20,45,68,77,64]
y=[46,68,89,34,45,24]
plt.scatter(x,y,color='red')
plt.xlabel('age')
plt.ylabel('salary')
plt.title('age vs salary chart')
```

### 2D Graph can plot on numpy array and list :


```python
product_id=[1,2,3,4,5,6,7,8]
online_price=[255,657,578,576,587,765,687,498]
offline_price=[280,700,600,650,800,700,500,700]
blackbazar_price=[260,780,690,640,880,720,540,730]
plt.plot(product_id,online_price,color='red',marker='D',label='online_price')
plt.plot(product_id,offline_price,color='green',marker='D',label='offline_price')
plt.plot(product_id,blackbazar_price,color='blue',marker='D',label='blackbazar_price')
plt.xlabel('product_id')
plt.ylabel('online_price')
plt.title('product_id vs price chart')
plt.legend()
plt.show()
```

## Scatter plot using array

```python
x=([10,20,45,68,77,64])
y=([46,68,89,34,45,24])
plt.scatter(x,y,color='blue')
plt.xlabel('rank')
plt.ylabel('salary')
plt.title('rank vs salary chart')
plt.show()
```

```python
year=[2010,2011,2012,2013,2014,2015,2015]
GDP=[4.6,4.8,4.9,5.4,5.4,6.2,6.5]
plt.scatter(year,GDP, color='red')
plt.plot(year,GDP, color='red')
plt.xlabel('year')
plt.ylabel('GDP')
plt.title('year vs GDP chart')
plt.show()
```

## Bar chart


```python
year=[2010,2011,2012,2013,2014,2015,2015]
GDP=[4.6,4.8,4.9,5.4,5.4,6.2,6.5]
plt.bar(year,GDP, color='green')
plt.xlabel('year')
plt.ylabel('GDP')
plt.title('year vs GDP chart')
plt.show()
```

```python
year=[2010,2011,2012,2013,2014,2015,2015]
GDP=[4.6,4.8,4.9,5.4,5.4,6.2,6.5]
plt.pie(year,labels=GDP,autopct="%0.1f%%")
plt.show()
```
